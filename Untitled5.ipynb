```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'*')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'3')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'>')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'r')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'b')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'o:r')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'o:b',ms=20)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'o:b',ms=99)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'o',ms=99,mec='g')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
plt.plot(x,y,'o',ms=99,mec='r',mfc='r')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```
